# Data Visualization with Matplotlib

Matplotlib is a powerful plotting library for Python that enables the creation of static, animated, and interactive visualizations. It is widely used for data visualization in data science and analytics.

```python
%pip install matplotlib
```

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[1,4,5,7,8]

## create a line plot
plt.plot(x,y)
plt.xlabel('X label')
plt.ylabel('y label')
plt.title('Basic Line Plot')
plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,4,5,7,8]

## create customized line plot

plt.plot(x, y, color="red", linestyle=":", marker="*", linewidth=1, markersize=9)
plt.grid(True)
```

```python
## Multiplot
## Sample Data
x = [1,2,3,4,5]
y1 = [1, 4, 9, 16, 25]
y2 = [1, 2, 3, 4, 5]

plt.figure(figsize=(9,5))

## parameters(how many rows, how many columns, in which should it take place)
plt.subplot(2,2,1)
plt.plot(x, y1, color="green")
plt.title("Plot 1")

plt.subplot(2,2,2)
plt.plot(x, y2, color="red")
plt.title("Plot 1")
```

```python
###Barplot
categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 6, 3, 8, 9]
plt.bar(categories, values, color="purple")
```

Histogram are used to represent the distribution of a dataset. They divide the data into bins and count the number of data points in each one

```python
data = [1, 3, 6, 3, 5, 3, 3, 4, 5, 3, 5, 7, 7, 9]

#create histogram
plt.hist(data, bins=5, color="orange", edgecolor="black")
```

```python
#create a scatter plot
#sample data
x = [1, 2, 3, 4, 5]
y = [2, 3, 4, 5, 6]

plt.scatter(x, y, color="blue", marker="*")
```

```python
## pie chart
labels = ['A', 'B', 'C', 'D']
sizes = [30, 20, 40, 10]
color = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1,0,0,0)

## create
plt.pie(sizes, explode=explode, labels=labels, colors=color, autopct="%1.1f%%", shadow=True)
```
